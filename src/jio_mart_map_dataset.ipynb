# Import Packages

```python
import pandas as pd

from utils import get_labels
from constants import JIO_MART_DATASET_PATH, GPC_PATH
```

```python
df_gpc = pd.read_csv(GPC_PATH)
```

```python
df_jio = pd.read_csv(JIO_MART_DATASET_PATH)
```

```python
df_jio
```

```python
len(df_jio[df_jio["Class"].isna()])
```

```python
df_jio.dropna(subset=["Class"], inplace=True)
```

```python
df_jio
```

```python
df_jio = df_jio[["product name", "Segment", "Family", "Class"]]
```

```python
df_jio["Segment"] = df_jio["Class"].apply(lambda x: df_gpc[df_gpc["ClassTitle"]==x]["SegmentTitle"].tolist()[0])
df_jio["Family"] = df_jio["Class"].apply(lambda x: df_gpc[df_gpc["ClassTitle"]==x]["FamilyTitle"].tolist()[0])
```

```python
df_jio.rename(columns={
    "product name": "product_name",
    "Segment": "segment",
    "Family": "family",
    "Class": "class"
}, inplace=True)
```

```python
classes = df_jio["class"].unique().tolist()
true_classes = df_gpc["ClassTitle"].unique().tolist()
get_labels(classes, true_classes)["incorrect_labels"]
```

```python
segments = df_jio["segment"].unique().tolist()
true_segments = df_gpc["SegmentTitle"].unique().tolist()
get_labels(segments, true_segments)["incorrect_labels"]
```

```python
families = df_jio["family"].unique().tolist()
true_families = df_gpc["FamilyTitle"].unique().tolist()
get_labels(families, true_families)["incorrect_labels"]
```

```python
df_jio.to_csv(JIO_MART_DATASET_PATH, index=False, encoding="utf-8-sig")
```
